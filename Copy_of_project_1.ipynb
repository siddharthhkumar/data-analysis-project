```python
# 1. Import magic tools
import pandas as pd
import matplotlib.pyplot as plt

# 2. Create simple data (like a tiny Excel sheet)
data = {
    'Fruit': ['Apple', 'Banana', 'Orange'],
    'Sales': [45, 33, 28]
}
df = pd.DataFrame(data)

# 3. Show your data
print("My First Data Table:")
display(df)
```

```python
# 4. Create a colorful bar chart
df.plot(kind='bar', x='Fruit', y='Sales', color=['red', 'yellow', 'orange'])
plt.title("Fruit Sales")
plt.show()
```

```python
store_data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Customers': [120, 85, 93, 110, 150],
    'Revenue ($)': [540, 320, 410, 590, 820]
}
```

# Task
Analyze and visualize the provided store data.

## Convert data to dataframe

### Subtask:
Convert the `store_data` dictionary to a pandas DataFrame.


**Reasoning**:
Convert the `store_data` dictionary into a pandas DataFrame and display the first few rows to verify the conversion.



```python
store_df = pd.DataFrame(store_data)
display(store_df)
```

## Analyze data

### Subtask:
Calculate and display some basic statistics for the 'Customers' and 'Revenue ($)' columns.


**Reasoning**:
Calculate and display descriptive statistics for the 'Customers' and 'Revenue ($)' columns of the `store_df` DataFrame.



```python
selected_columns = store_df[['Customers', 'Revenue ($)']]
statistics = selected_columns.describe()
display(statistics)
```

## Visualize data

### Subtask:
Create two plots: a bar chart for 'Customers' by 'Day' and a line plot for 'Revenue ($)' by 'Day'.


**Reasoning**:
Create a bar chart for 'Customers' by 'Day' and a line plot for 'Revenue ($)' by 'Day' using the store_df DataFrame, and display both plots.



```python
store_df.plot(kind='bar', x='Day', y='Customers')
plt.title("Daily Customers")
plt.show()

store_df.plot(kind='line', x='Day', y='Revenue ($)')
plt.title("Daily Revenue ($)")
plt.show()
```

## Summary:

### Data Analysis Key Findings

* The data was successfully converted into a pandas DataFrame with columns 'Day', 'Customers', and 'Revenue (\$)', ready for analysis.
* Basic statistics were calculated, showing the count, mean, standard deviation, minimum, maximum, and quartiles for both 'Customers' and 'Revenue (\$)' columns.
* A bar chart visualizing daily customer counts and a line plot visualizing daily revenue were successfully generated and displayed.
